```python
# подключение библиотек
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
from scipy.optimize import fsolve
```

```python
v1 = [10, 10,  10]
v2 = [ 0,  0, -10]
def get_nd_vector_length(n_vector):
    sum = 0
    for i in range(len(n_vector)):
        sum += n_vector[i]*n_vector[i]
    return sqrt(sum)

print(f'длинна вектора {v1} равна {get_nd_vector_length(v1)}')
print(f'длинна вектора {v2} равна {get_nd_vector_length(v2)}')
```

```python
a,b,c,d,e = 5.0, 3.0, 7.0, 1.0, 2.0
x = np.arange(-8, 8, 0.001)
y1_c, y2_c, y1_e, y2_e, y1_h, y2_h = [],[],[],[],[],[]

y1_c = [a*sqrt(1-v*v/(a*a)) if v*v/(a*a) <= 1.0 else None for v in x]
y2_c = [-i if i is not None else None for i in y1_c]
y1_e = [c*sqrt(1-v*v/(b*b)) if v*v/(b*b) <= 1.0 else None for v in x]
y2_e = [-i if i is not None else None for i in y1_e]
y1_h = [e*sqrt(v*v/(d*d)-1) if v*v/(d*d) >= 1.0 else None for v in x]
y2_h = [-i if i is not None else None for i in y1_h]

plt.plot(x,y1_c,x,y2_c,x,y1_e,x,y2_e,x,y1_h,x,y2_h)

plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax = Axes3D(fig)
X = np.arange(-11, 11, 0.05)
Y = np.arange(-11, 11, 0.05)
X, Y = np.meshgrid(X, Y)

Ze1 =  30 - (X+5)**2 - (Y+5)**2
Ze2 =  30 - (X+5)**2 - (Y+5)**2
Zh1 = -15 + (X-5)**2 + (Y-5)**2
Zh2 = -15 + (X-5)**2 + (Y-5)**2

for r,ve in enumerate(Ze1):
    for c,va in enumerate(ve):
        if va >= 0:
            Ze1[r,c] = np.sqrt(va)
            Ze2[r,c] = -Ze1[r,c]
        else:
            Ze1[r,c] = np.nan
            Ze2[r,c] = np.nan
for r,ve in enumerate(Zh1):
    for c,va in enumerate(ve):
        if va >= 0:
            Zh1[r,c] = np.sqrt(va)
            Zh2[r,c] = -Zh1[r,c]
        else:
            Zh1[r,c] = np.nan
            Zh2[r,c] = np.nan
        
ax.plot_surface(X, Y, Zh2)
ax.plot_surface(X, Y, Ze2)
ax.plot_surface(X, Y, Ze1)
ax.plot_surface(X, Y, Zh1)

show()

# ==================================================================================== 
#   конкретно для элипсойда и однополосного гиперболойда решение не очень красивое:
#   рендеринг не дружит с отсутствующими точками
# ====================================================================================
```

```python
def f1(x):
    return x**2-1

def f2(x):
    return (np.exp(x) + x - 1)/x
    
x = np.linspace(-3, 5, 101)
plt.plot(x, f1(x))
plt.plot(x, f2(x))

def equation(point):
    xx,yy = point
    return (-yy+xx**2-1, -1+np.exp(xx) + xx*(1-yy))

print("Корни уравнения:")
x1,y1 = fsolve(equation, (-2,3))
x2,y2 = fsolve(equation, ( 3,5))
x3,y3 = fsolve(equation, (-4,15))
print(f"1) {x1},{y1}")
print(f"2) {x2},{y2}")
print(f"3) {x3},{y3}")
```

```python
# подставим y из первого уравнения во второе неравенство

#функция монотонно возрастающая
def ff(xx):
    return np.exp(xx) + xx**3 - 1

x = np.linspace(-2, 2, 199)
plt.plot(x, ff(x))
plt.grid(True) 

# подойдем справа
x1r = fsolve(ff, 1)
y1r = ff(x1r)
print(x1r, y1r)
# подойдем слева
x1l = fsolve(ff, -1)
y1l = ff(x1l)
print(x1l, y1l)
# возьмем среднее (если хочется)
x1 = (x1r+x1l)/2
y1 = ff(x1)
print(x1, y1)
```
