```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5*np.pi, 5*np.pi, 201)
k1 = 1
k2 = 1.2
plt.plot(x, np.cos(k1*x))
plt.plot(x, np.cos(k2*x))
plt.xlabel('x')
plt.ylabel('y(x)')
plt.show()
```
